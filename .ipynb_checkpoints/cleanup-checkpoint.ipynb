```python
import pandas as pd
```

```python
#set dataframes for cleanup
characters = pd.read_csv("data/simpsons_characters_edit.csv")
lines = pd.read_csv("data/simpsons_script_lines.csv", error_bad_lines=False)
episodes = pd.read_csv("data/simpsons_episodes.csv")
```

```python
#including columns needed for analysis
episodes = episodes[["id", "season", "number_in_season", "number_in_series", "title", "imdb_rating", "image_url"]]
episodes = episodes[episodes.season != 28]
```

```python
null_columns=episodes.columns[episodes.isnull().any()]
episodes[null_columns].isnull().sum()

#search for null in columns
print(episodes[episodes.isnull().any(axis=1)][null_columns])
#clear NAN columns
episodes.iloc[189,6] = ""
episodes.count()
```

```python
#rename columns in episodes
episodes = episodes.rename(columns={"id": "episode_id", "number_in_season": "season_episode_no", "number_in_series": "episode_overall_no"})
episodes.sort_values(by=['episode_id'], inplace=True)
episodes = episodes.set_index("episode_id")
```

```python
#set columns and rows for characters
characters = characters.rename(columns={"id":"char_id"})
characters.sort_values(by=["char_id"], inplace=True)
characters = characters.set_index("char_id")
characters
```

```python
#lots of internal dialogue in simpsons, deleting for actual lines
lines = lines.loc[lines["speaking_line"]==True]
lines = lines[["id", "episode_id", "number", "character_id", "normalized_text", "word_count"]]
lines = lines.rename(columns={"id":"line_id"})
type(lines)
```

```python
#clean up nan rows, and sort and reindex
lines = lines.dropna()
lines.sort_values(by=["line_id"], inplace=True)
lines = lines.set_index("line_id")
type(lines)


```

```python
lines.count()
```

```python
#export csv and json files for future import
episodes.to_csv("data/episodes_final.csv")
lines.to_csv("data/lines_final.csv")
characters.to_csv("data/characters_final.csv")
episodes.to_json("data/episodes_final.json")
lines.to_json("data/lines_final.json")
characters.to_json("data/characters_final.json")
```
