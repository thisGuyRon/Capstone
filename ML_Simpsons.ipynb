```python
from sqlalchemy import create_engine
import pandas as pd
#import matplotlib.pyplot as plt
# PyMySQL 
import pymysql
import config
pymysql.install_as_MySQLdb()
```

```python
#connection to database
engine = create_engine("mysql://root:" + config.root_pwd + "@localhost/simpsons")
conn = engine.connect()
```

```python
#pull current springfield dataset
#pulling top 40 most prevalent in episodes
ml1_df = pd.read_sql('select * from m_learning1', conn)
ml1_df
```

```python
# Assign X (data) and y (target)
X = ml1_df.drop("gender", axis=1)
y = ml1_df["gender"]
print(X.shape, y.shape)
```

```python
#Split our data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=33, stratify=y)
```

```python
#Create a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
#fit the data with train data
classifier.fit(X_train,y_train)
```

```python
#Validate the model using the test data
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
#make predictions
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
#print comparison dataframe
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```
